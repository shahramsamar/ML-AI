```python
n = 1
while n <= 5:
    print(n, end=',')
    n += 1
```

```python
n = 30
while n > 20:
    print(n)
    n -= 2
```

```python
x = [10, 12, 17]
s = 0
i = 0
while i < 3:
    s += x[i]
    i += 1
print(s)    
```

```python
a, b = 1, 1
while a < 150:
    print(a, end=',')
    a, b = b, a+b
```
