```python
def f(x):
    return(x ** 2 + 1)
result = f(3)
print(result)
```

```python
def f(a, b):
    return a + 2 * b
result = f(3, 2)
print(result)
```

```python
# Passing Arguments with key and Value
result = f(b=3, a=2)
print(result)
```

```python
# function with default parameters
def age(a, b=2026):
    return b - a
result = age(1989)
print(result)
```

```python
# arbitrary number of arguments
def f(*x):
    t = 0
    for n in x:
        t += n
    return t
print(f(2, 3, 5, 20, 50))    
        
```

```python
# passing list as argument
def mean(lst):
    return sum(lst) / len(lst)
result = mean([1, 2, 3, 4])  
print(result)
```

```python
# passing dictionary as argument
def f(d):
    d['age'] += 1
mydict = {'name':'shahram', 'age':36}
f(mydict)
print(mydict)
```
