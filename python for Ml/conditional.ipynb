```python
n = 1
match n:
    case 1: print("Saturday")
    case 2: print("Sunday")
    case 3: print("Monday")
    case 4: print("Tuesday")
    case 5: print("Wednesday")
    case 6: print("Thursday")
    case 7: print("Friday")
    case _: print("Invalid day number")
```

```python
user = "manager"
match  user:
    case "admin" | "manager":
        print("Full Access")
    case "Guest":
        print("Limited access")
    case _:
        print("No access")
        
        
        
```
